## Pos/Neg/Zero

````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'plusMinus' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def plusMinus(arr):
    # Write your code here
    
    """
    We can either do it sequentially:
    calc sums, then divide
    
    But its probably better to go the efficient
    route where we do it all in one pass, so its
    more vectorizeable.
    
    For each element in the arr:
        if / else: sign -> act
        
    Ineffective...
    
    We want 6 digits after decimal, so 
    we'll use round or :.6f
    
    There is no way around division.
    
    We could filter to avoid if/else, but that
    will cause separate loops. 
    
    I make the assumption that its better to 
    loop multiple times than to use if/else branching.
    
    Complexity:
    - time: O(n^2) due to sum of filter
    - auxilirary space: O(1)
    - input space: O(n)
    """
    
    n = len(arr)
    
    pos = sum(1 for _ in filter(lambda item: item > 0, arr)) / n
    neg = sum(1 for _ in filter(lambda item: item < 0, arr)) / n
    zero = sum(1 for _ in filter(lambda item: item == 0, arr)) / n
    print(f"{pos:.6f}\n{neg:.6f}\n{zero:.6f}")

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    plusMinus(arr)
````

#### The Anatomy of `sum(1 for _ in filter(lambda x: x > 0, arr))`

A single line of `sum(1 for _ in filter(...))` is **neither a nested loop nor two separate complete loops.** It is a **lazy-evaluated data pipeline**. It results in exactly one single pass over the array. Let's break down exactly what happens in memory and CPU.

If this were two separate loops, Python would filter the whole array, create a new temporary list in memory, and then loop over that new list to sum it. **That does not happen here.** Instead, Python uses iterators and generators, which work like an assembly line:

1. **`filter(...)` (The Gatekeeper):** This does not execute immediately. It just sets up an iterator. It waits to be asked for a value.
2. **`1 for _ in ...` (The Transformer):** This is a generator expression. It asks the filter for a value, and if it gets one, it turns it into a `1`. It also waits to be asked.
3. **`sum(...)` (The Consumer):** This is the boss. It drives the whole process.

**The Execution Flow (Step-by-Step):**

* `sum` says: *"I need a number to add to my total."*
* The generator says: *"Hey `filter`, give me a valid item so I can turn it into a 1."*
* `filter` looks at `arr[0]`. Is it `> 0`?
* If **Yes**: `filter` passes it to the generator, the generator yields `1`, and `sum` adds it to the total.
* If **No**: `filter` skips it and looks at `arr[1]` without bothering the generator or `sum`.

## Time conversion

````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    # Write your code here
    
    """
    Its simple string manipulation
    
    We just split at the colon.
    
    Wether or not we modify the time (hours)
    depends on both AM/PM and current time.
    
    The string mutation is, I think, O(N) time 
    for the split method and O(N) space where 
    N is decided by string length.
    """
    
    is_am = "AM" in s
    is_pm = "PM" in s
    if not (is_am or is_pm):
        raise ValueError("Malformed string: missing AM or PM")
        
    s = s.strip("AM").strip("PM")

    try:
        hh, mm, ss = map(lambda item: int(item), s.split(":"))
    except Exception as e:
        raise ValueError("Malformed string: invalid value for hours, minutes or seconds")
    
    if is_am and hh >= 12:
        hh -= 12
    elif is_pm and hh < 12:
        hh += 12
        
    return f"{str(hh).zfill(2)}:{str(mm).zfill(2)}:{str(ss).zfill(2)}"


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
````

```python
def timeConversion(s: str) -> str:
    # O(1) Time, O(1) Space 
    # Guaranteed 10-char strict format: "hh:mm:ssAM"
    
    period = s[-2:] # Extract "AM" or "PM"
    time_str = s[:-2] # Extract "hh:mm:ss"
    
    # map(int, ...) is faster and cleaner than lambdas
    hh, mm, ss = map(int, time_str.split(':'))
    
    if period == "AM" and hh == 12:
        hh = 0
    elif period == "PM" and hh != 12:
        hh += 12
        
    # Elite execution: native f-string zero-padding (:02d)
    # This bypasses the need to convert back to strings and call .zfill()
    return f"{hh:02d}:{mm:02d}:{ss:02d}"

```

##### Key Upgrades:

1. **`s[-2:]` and `s[:-2]**`: This is strict $O(1)$ pointer math in C (under the Python hood), far safer and faster than `.strip()`.
2. **`map(int, ...)`**: Clean, functional mapping.
3. **`f"{hh:02d}"`**: The `:02d` instruction tells Python's C-backend to format the integer to 2 digits, zero-padded. This is highly optimized and standard for modern Python data pipelines.
